```python
import ipywidgets as widgets
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import pandas as pd
from IPython.display import display, HTML

uniform_layout = widgets.Layout(width='98%')

# Input for clicks per ring
clicks_per_ring_input = widgets.FloatText(
    value=1.840,
    description='Clicks per Ring:',
    style={'description_width': '150px'},
    layout=uniform_layout
)

# Input for ring radius and included rings
ring_radius_input = widgets.FloatText(
    value=2.0,
    description='Ring Radius (cm):',
    style={'description_width': '150px'},
    layout=uniform_layout
)

lowest_ring_input = widgets.IntSlider(
    value=6,
    min=1,
    max=10,
    step=1,
    description='Lowest Ring:',
    style={'description_width': '150px'},
    layout=uniform_layout
)

# Slider for spread factor
spread_slider = widgets.FloatSlider(
    value=4.0,
    min=0.5,
    max=10.0,
    step=0.1,
    description='Spread:',
    style={'description_width': '150px'},
    layout=uniform_layout
)

# Input for arrow diameter
arrow_diameter_input = widgets.FloatText(
    value=6.0,
    description='Arrow Diameter (mm):',
    style={'description_width': '150px'},
    layout=uniform_layout
)

# Switch for handedness
handedness_switch = widgets.ToggleButtons(
    options=[('Right-Handed', 'right'), ('Left-Handed', 'left')],
    value='right',
    style={'button_width': '140px'},
    layout=widgets.Layout(
        display='flex',
        justify_content='center',
        align_items='center',
    )
)

# Output areas for the target plot and results
output_graph = widgets.Output()
output_results = widgets.Output()
round_results = widgets.Output()

# Generate and Check buttons
generate_button = widgets.Button(
    description="Generate Arrows",
    button_style="primary"
)

check_button = widgets.Button(
    description="Check",
    button_style="success",
    layout=uniform_layout
)

next_round_button = widgets.Button(
    description="Next Round",
    button_style="warning",
    layout=uniform_layout
)

start_round_button = widgets.Button(
    description="Start Round Mode",
    button_style="warning",
    layout=uniform_layout
)

# Variables to store arrow data
arrow_positions = []
calculated_center = (0, 0)
arrow_ring_values = []
corrected_positions = []
new_center = (0, 0)

# DataFrame to store round results
round_data = pd.DataFrame(
    columns=["Set", "Arrow", "Original X", "Original Y", "Corrected X", "Corrected Y", "Original Score", "Corrected Score", "Distance"]
)
current_set = 1
is_round_mode = False

# Clear outputs when settings are changed
def clear_outputs(change):
    output_graph.clear_output()
    output_results.clear_output()
    round_results.clear_output()

handedness_switch.observe(clear_outputs, names='value')
spread_slider.observe(clear_outputs, names='value')
arrow_diameter_input.observe(clear_outputs, names='value')
lowest_ring_input.observe(clear_outputs, names='value')

# Generate the target and arrows
def generate_target(Check=False):
    global arrow_positions, calculated_center, arrow_ring_values, corrected_positions, new_center
    hor_correction_input.value = 0
    vert_correction_input.value = 0
    with output_graph:
        output_graph.clear_output()

        # Get user-defined parameters
        ring_radius = ring_radius_input.value
        lowest_ring = lowest_ring_input.value
        spread_factor = spread_slider.value
        arrow_diameter = arrow_diameter_input.value / 10.0  # Convert mm to cm

        # Create the figure and axes
        fig, ax = plt.subplots(figsize=(6, 6))

        largest_ring_radius = (11-lowest_ring) * ring_radius
        ax.set_xlim(-largest_ring_radius-1, largest_ring_radius+1)
        ax.set_ylim(-largest_ring_radius-1, largest_ring_radius+1)
        ax.set_aspect('equal', adjustable='datalim')

        # Define colors for rings
        colors = ['#ffffff', '#000000', '#00B4E4', '#F65058', '#FFE552']
        for i in range(lowest_ring, 11, 1):
            color = colors[(i - 1) // 2]
            edge_color = "black" if not i == 4 else "white"
            line_width = 1 if i % 2 == 0 else 0.5
            circle = plt.Circle((0, 0), (11-i)* ring_radius, color=color, ec=edge_color, fill=True,
                                alpha=1, linewidth=line_width)
            ax.add_artist(circle)

        if not Check:
            # Generate 3 arrow positions randomly around some random center
            arrow_center = (
                np.random.uniform(-ring_radius * lowest_ring / 2, ring_radius * lowest_ring / 2),
                np.random.uniform(-ring_radius * lowest_ring / 2, ring_radius * lowest_ring / 2)
            )
            arrow_positions = [
                (arrow_center[0] + np.random.normal(0, ring_radius / 4 * spread_factor),
                 arrow_center[1] + np.random.normal(0, ring_radius / 4 * spread_factor))
                for _ in range(3)
            ]

            # Calculate the average (x, y) of these positions
            calculated_center = (
                np.mean([pos[0] for pos in arrow_positions]),
                np.mean([pos[1] for pos in arrow_positions])
            )

            # Calculate ring values for each arrow
            arrow_ring_values = []
            for x, y in arrow_positions:
                distance = np.sqrt(x**2 + y**2)
                ring_score = 0  # Default to 0 for scores outside the lowest ring
                for i in range(10, lowest_ring - 1, -1):  # Only check up to the lowest ring
                    if distance <= ((11 - i) * ring_radius + arrow_diameter / 2):
                        ring_score = i
                        break
                arrow_ring_values.append(ring_score)

        original_plotted = False
        corrected_plotted = False
        # Plot the arrows using circles to incorporate size
        for x, y in arrow_positions:
            if not original_plotted:
                arrow_circle = patches.Circle((x, y), radius=arrow_diameter/2, color='black', alpha=0.8, label='Original Arrows')
                original_plotted = True
            else:
                arrow_circle = patches.Circle((x, y), radius=arrow_diameter/2, color='black', alpha=0.8)
            ax.add_patch(arrow_circle)

        if Check:
            # Plot corrected positions and centers
            for x, y in corrected_positions:
                if not corrected_plotted:
                    corrected_circle = patches.Circle((x, y), radius=arrow_diameter / 2, color='lime', alpha=0.8, label='Corrected Arrows')
                    corrected_plotted = True
                else:
                    corrected_circle = patches.Circle((x, y), radius=arrow_diameter / 2, color='lime', alpha=0.8)
                ax.add_patch(corrected_circle)
            ax.plot(calculated_center[0], calculated_center[1], 'rx', markersize=12, mew=2, label='Original Group Center')
            ax.plot(new_center[0], new_center[1], 'gx', markersize=12, mew=2, label='Corrected Group Center')

        # Add center cross
        ax.plot(0, 0, 'k+', markersize=10, mew=2)
        if handedness_switch.value == 'left':
            ax.invert_xaxis()
        ax.set_xlabel('Out/In' if handedness_switch.value == 'right' else 'In/Out')
        ax.set_ylabel('Up/Down')
        #ax.legend(loc='upper left', bbox_to_anchor=(1.05, 1), borderaxespad=0.)
        plt.title("Target with Arrow Group")
        plt.show()

# Check corrections and store results in the DataFrame
def check_correction(change):
    global corrected_positions, new_center, round_data, current_set
    ring_radius = ring_radius_input.value
    lowest_ring = lowest_ring_input.value
    with output_results:
        output_results.clear_output()
        try:
            hor_corr = float(hor_correction_input.value)
            vert_corr = float(vert_correction_input.value)
            click_value = float(clicks_per_ring_input.value)
        except ValueError:
            display(HTML("<p style='color:red;'>Invalid input for corrections.</p>"))
            return

        new_center = (calculated_center[0] + (hor_corr * click_value),
                      calculated_center[1] + (vert_corr * click_value))

        corrected_positions = [
            (x + (hor_corr * click_value), y + (vert_corr * click_value))
            for x, y in arrow_positions
        ]

        # Calculate new scores
        corrected_ring_values = []
        arrow_diameter = arrow_diameter_input.value / 10.0
        for x, y in corrected_positions:
            distance = np.sqrt(x**2 + y**2)
            ring_score = 0  # Default to 0 for scores outside the lowest ring
            for i in range(10, lowest_ring - 1, -1):  # Only check up to the lowest ring
                if distance <= ((11 - i) * ring_radius + arrow_diameter / 2):
                    ring_score = i
                    break
            corrected_ring_values.append(ring_score)

        generate_target(Check=True)

        # Calculate distance between centers
        distance_between_centers = np.sqrt((new_center[0] - calculated_center[0])**2 + (new_center[1] - calculated_center[1])**2)

        # Store results in DataFrame
        for i, (original, corrected) in enumerate(zip(arrow_positions, corrected_positions), start=1):
            round_data = pd.concat([
                round_data,
                pd.DataFrame({
                    "Set": [current_set],
                    "Arrow": [i],
                    "Original X": [original[0]],
                    "Original Y": [original[1]],
                    "Corrected X": [corrected[0]],
                    "Corrected Y": [corrected[1]],
                    "Original Score": [arrow_ring_values[i-1]],
                    "Corrected Score": [corrected_ring_values[i-1]],
                    "Distance": [distance_between_centers]
                })
            ], ignore_index=True)

        # Display round results after the set ends
        display(HTML(f"""
        <h3>Set {current_set} Results</h3>
        <p>Original Scores: {sorted(arrow_ring_values, reverse=True)} (Total: {sum(arrow_ring_values)})</p>
        <p>Corrected Scores: {sorted(corrected_ring_values, reverse=True)} (Total: {sum(corrected_ring_values)})</p>
        <p>Distance between original and new group centers: {distance_between_centers:.2f} cm</p>
        """))

# Next round function for round mode
def next_round(change):
    global current_set, is_round_mode
    if is_round_mode and current_set < 5:
        current_set += 1
        generate_target(Check=False)
    elif is_round_mode:
        with round_results:
            round_results.clear_output()
            # Prepare formatted tables
            formatted_summary = """<style>
                table {width: 100%; border-collapse: collapse; font-family: monospace; font-size: 16px;}
                th, td {border: 1px solid black; padding: 10px; text-align: center;}
                th {font-weight: bold;}
                .yellow {background-color: yellow;}
                .red {background-color: red; color: white;}
                .black {background-color: black; color: white;}
                .white {background-color: white;}
            </style>
            <table>
            <tr><th>Set</th><th>Original Scores</th><th>Sum</th><th>Corrected Scores</th><th>Sum</th><th>Difference</th><th>Running Total</th></tr>
            """

            running_total = 0

            for set_num in range(1, current_set + 1):
                original_scores = list(round_data[round_data['Set'] == set_num]['Original Score'])
                corrected_scores = list(round_data[round_data['Set'] == set_num]['Corrected Score'])
                original_sum = sum(original_scores)
                corrected_sum = sum(corrected_scores)
                diff = corrected_sum - original_sum
                running_total += corrected_sum

                def colorize(score):
                    if score == 10 or score == 9:
                        return f"<span class='yellow'>{score}</span>"
                    elif score == 8 or score == 7:
                        return f"<span class='red'>{score}</span>"
                    elif score == 6 or score == 5:
                        return f"<span class='red'>{score}</span>"
                    elif score == 4 or score == 3:
                        return f"<span class='black'>{score}</span>"
                    elif score <= 2:
                        return f"<span class='white'>{score}</span>"
                    return str(score)

                original_scores_colored = [colorize(s) for s in original_scores]
                corrected_scores_colored = [colorize(s) for s in corrected_scores]

                formatted_summary += f"<tr><td>{set_num}</td><td>{' '.join(original_scores_colored)}</td>"
                formatted_summary += f"<td><strong>{original_sum}</strong></td><td>{' '.join(corrected_scores_colored)}</td>"
                formatted_summary += f"<td><strong>{corrected_sum}</strong></td><td>{diff:+}</td><td>{running_total}</td></tr>"

            total_original = sum(round_data['Original Score'])
            total_corrected = sum(round_data['Corrected Score'])
            total_diff = total_corrected - total_original

            formatted_summary += f"<tr><th>Total</th><th></th><th><strong>{total_original}</strong></th>"
            formatted_summary += f"<th></th><th><strong>{total_corrected}</strong></th><th>{total_diff:+}</th><th>{running_total}</th></tr>"
            formatted_summary += "</table>"

            display(HTML(formatted_summary))

# Start a 10-round session
def start_round_mode(change):
    global current_set, round_data, is_round_mode
    round_data = pd.DataFrame(
        columns=["Set", "Arrow", "Original X", "Original Y", "Corrected X", "Corrected Y", "Original Score", "Corrected Score", "Distance"]
    )
    current_set = 1
    is_round_mode = True
    generate_target(Check=False)

# Inputs for horizontal and vertical correction
hor_correction_input = widgets.FloatText(
    value=0.0,
    description='Horizontal clicks:',
    style={'description_width': '150px'},
    layout=uniform_layout
)

vert_correction_input = widgets.FloatText(
    value=0.0,
    description='Vertical clicks:',
    style={'description_width': '150px'},
    layout=uniform_layout
)

hor_correction_buttons = widgets.HBox([
    widgets.Button(description='Out', button_style='info', layout=widgets.Layout(width='49%')),
    widgets.Button(description='In', button_style='info', layout=widgets.Layout(width='48%'))
], layout=widgets.Layout(width='100%'))

for btn in hor_correction_buttons.children:
    btn.on_click(adjust_horizontal)

vert_correction_buttons = widgets.VBox([
    widgets.Button(description='Up', button_style='info', layout=uniform_layout),
    widgets.Button(description='Down', button_style='info', layout=uniform_layout)
], layout=widgets.Layout(align_items='center'))

def adjust_vertical(change):
    button = change
    if button.description == 'Up':
        vert_correction_input.value -= 1
    elif button.description == 'Down':
        vert_correction_input.value += 1

for btn in vert_correction_buttons.children:
    btn.on_click(adjust_vertical)

# Connect button clicks to functions
generate_button.on_click(lambda change: generate_target(Check=False))
check_button.on_click(check_correction)
next_round_button.on_click(next_round)
start_round_button.on_click(start_round_mode)


# Organize the layout
settings_box = widgets.VBox([
    widgets.HTML("<h2>Settings</h2>"),
    clicks_per_ring_input,
    ring_radius_input,
    lowest_ring_input,
    spread_slider,
    arrow_diameter_input,
    handedness_switch
], layout=widgets.Layout(border='1px solid black', padding='10px', width='32%'))

sight_controls_box = widgets.VBox([
    widgets.HTML("<h2>Sight Controls</h2>"),
    vert_correction_buttons,
    vert_correction_input,
    hor_correction_buttons,
    hor_correction_input,
    check_button
], layout=widgets.Layout(border='1px solid black', padding='10px', width='32%'))

actions_box = widgets.VBox([
    widgets.HTML("<h2>Actions</h2>"),
    generate_button,
    start_round_button,
    next_round_button
], layout=widgets.Layout(border='1px solid black', padding='10px', width='32%'))

layout = widgets.HBox([
    settings_box,
    sight_controls_box,
    actions_box
], layout=widgets.Layout(justify_content='space-between'))

# Display the layout
display(layout, output_graph, output_results, round_results)
```
